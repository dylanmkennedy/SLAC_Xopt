# Automatic emittance measurement example
In this case we have 4 quadrupoles, 3 that are used to minimize the incoming beam and
 the fourth to perform the quad scan measurement. The nominal x/y emittances are 0.1
 mm-mrad. See scripts package for algorithm details.

```python
from test_problems.quad_quadlet.src import VARIABLES, beamsize_quadlet
from scripts.optimize_function import optimize_function
from scripts.characterize_emittance import characterize_emittance
```

## Optimize upstream beam parameters
The goal is to minimize the beamsize at zero quad measurement strength (`k4 = 0`).

```python
from xopt import VOCS

opt_vocs = VOCS(
        variables = {"k1":[0,100], "k2":[-100,0], "k3":[0,100]},
        constants = {"k4":0.0},
        objectives = {"total_size": "MINIMIZE"}
    )
opt_x = optimize_function(
    opt_vocs, beamsize_quadlet, n_iterations=30,
)
```

```python
opt_x.data
```

## Automatic quad scan and emittance characterization

```python
def eval_beamsize(input_dict):
    beamsize_results = beamsize_quadlet(input_dict)
    beamsize_results["rms_x"] = (beamsize_results["m11"] ** 0.5) * 1e3
    beamsize_results["rms_y"] = (beamsize_results["m22"] ** 0.5) * 1e3

    return beamsize_results

emit_vocs = exp_vocs = VOCS(
        variables = {"k4":[-300,300]},
        objectives = {"rms_x":"MINIMIZE"},
        constants = opt_x.data.iloc[-1][["k1","k2","k3"]].to_dict()
    )

quad_length = 0.01
drift_length = 0.5

emit_results, emit_X = characterize_emittance(
    emit_vocs,
    eval_beamsize,
    quad_length,
    drift_length,
    quad_strength_key="k4",
    rms_x_key="rms_x",
    rms_y_key="rms_y",
    quad_scan_analysis_kwargs={"visualize": True}
)
```
